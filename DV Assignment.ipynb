```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('Project.csv')
data
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='product_category', y='transaction_qty', data=data)
plt.title('Transaction Quantities by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Transaction Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month Name', y='Total_Bill', data=data.groupby('Month Name')['Total_Bill'].sum().reset_index())
plt.title('Total Bills by Month')
plt.xlabel('Month')
plt.ylabel('Total Bill')
plt.xticks(rotation=45)
plt.show()
```

```python
x = data['transaction_qty']  
y = data['unit_price']       
sizes = data['Total_Bill'] 
plt.figure(figsize=(10, 8))
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title('Bubble Chart of Transaction Quantity vs. Unit Price')
plt.xlabel('Transaction Quantity')
plt.ylabel('Unit Price')
plt.colorbar(label='Total Bill')

plt.show()
```

```python
column_to_plot = 'Total_Bill'
plt.figure(figsize=(10, 6))
plt.hist(data[column_to_plot], bins=30, color='skyblue', edgecolor='black')

plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

plt.show()
```

```python

x_column = 'transaction_qty'
y_column = 'Total_Bill'

plt.figure(figsize=(10, 6))
plt.scatter(data[x_column], data[y_column], alpha=0.5)

plt.title('Scatter Plot of Transaction Quantity vs. Total Bill')
plt.xlabel('Transaction Quantity')
plt.ylabel('Total Bill')

plt.show()
```

```python
column_to_plot = 'Total_Bill'
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[column_to_plot])
plt.title('Box Plot of Total Bill')
plt.xlabel('Total Bill')

plt.show()
```

```python
column_to_plot = 'Total_Bill'
plt.figure(figsize=(10, 6))
sns.violinplot(data=data[column_to_plot])
plt.title('Violin Plot of Total Bill')
plt.xlabel('Total Bill')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
column_to_plot = 'Total_Bill'
plt.figure(figsize=(10, 6))
sns.kdeplot(data[column_to_plot], shade=True)
plt.title('Density Plot of Total Bill')
plt.xlabel('Total Bill')

plt.show()
```
